### Typical 1. Total Time to Watch a Series

Compute total viewing time (no breaks) for a series:
- 5 seasons, 10 episodes per season
- Season 1: 45 minutes 35 seconds per episode
- Seasons 2–5: 35 minutes 15 seconds per episode
Display the total as hours, minutes, and seconds.

**Example Output:**
```
Total time: 39h 30m 0s
```


````python
# Typical 1: Total Time to Watch a Series

# Episode durations in seconds
season1_episode_seconds = 45 * 60 + 35 # Converting minutes to seconds
other_episode_seconds = 35 * 60 + 15

# Number of episodes & seasons
episodes_per_season = 10
seasons_2_5 = 4

# Calculate total seconds
total_seconds = (season1_episode_seconds * episodes_per_season +
                 other_episode_seconds * episodes_per_season * seasons_2_5) # 111950 seconds

# Converting total seconds to hours, minutes, seconds
hours = total_seconds // 3600
remaining_seconds = total_seconds % 3600
minutes = remaining_seconds // 60
seconds = remaining_seconds % 60

# Display Result
print("-" * 50)
print("\tTotal Series Watching Time!") 
print("-" * 50)
print(f"\t- Total time: {hours}h {minutes}m {seconds}s")
print("-" * 50)
````

### Typical 2. Savings Projection (3 Years)

Compute final savings after 3 years given:
- Save £10 every week for 3 years
- Year 1 bonus: +£20 each week
- Every 2 weeks: −£0.50 (sister)
Assume 52 weeks per year; use arithmetic and augmented assignment; no conditionals.

**Example Output:**
```
Final savings after 3 years: £4689.50
```


````python
#Typical 2: Savings Projection (3 Years)

# Given Data 
weekly_deposit = 10
total_years = 3
year1_bonus = 20
weekly_sister_amount = 0.50

weeks_per_year = 52


# Calculating total weeks
total_weeks = weeks_per_year * total_years  # 156 weeks

# Calculating total savings
year1_savings = year1_bonus * weeks_per_year  # 20 * 52 = 1040
weekly_savings_total = weekly_deposit * total_weeks # 10 * 156 = 1560

#Calculating Deductions
sister_deduction_total = weekly_sister_amount * (total_weeks // 2) # 0.50 * 78 = 39

#Total Savings (3 Years)
total_savings = 0 # Initialize savings to 0
# Using augmented assignment to add savings
total_savings += weekly_savings_total
total_savings += year1_savings
total_savings -= sister_deduction_total


# Displaying Result
print("-" * 50)
print("\tSavings Projection (3 Years)")
print("-" * 50)
print(f" - Final savings (3 years): £{total_savings:.2f}")
print("-" * 50)

````

### Stretch 3. Population Projection (5 Years)

Assume current population is 312032486 and 1 year = 365 days.
Rates: birth every 7s, death every 13s, immigrant every 45s.
Compute projected population for each of the next 5 years using integer division (`//`).

**Example Output:**
```
Year 1: 315,xxx,xxx
...
Year 5: ...
```


*Optional hint:* Compute yearly net change once; then add each year.


````python
# Stretch 3: Population Projection (5 Years)

# Given Data
current_population = 312032486

# Calculating 365 days in seconds
annual_year = 365 * 24 * 60 * 60  # annual_year = 31,536,000

# Events occurance rate in seconds (based on given data)
birth_rate = 7
death_rate = 13
immigrant_rate = 45

# Calculating yearly events
births = annual_year // birth_rate # 31,536,000 // 7
deaths = annual_year // death_rate # 31,536,000 // 13
immigrants = annual_year // immigrant_rate # 31,536,000 // 45

# Yearly Net Change
yearly_net_change = births + immigrants - deaths 

# Computing population for each year 
year1 = current_population + yearly_net_change
year2 = year1 + yearly_net_change
year3 = year2 + yearly_net_change
year4 = year3 + yearly_net_change
year5 = year4 + yearly_net_change
   

# Decorative Output of Population
print("=" * 50)
print("\tPOPULATION PROJECTION (5 Years)")
print("=" * 50)
print(f"- Current population: {current_population:,}")
print(f"- Total yearly change: {yearly_net_change:,}")
print("-" * 50)
print(f"Year 1: {year1:,}")
print(f"Year 2: {year2:,}")
print(f"Year 3: {year3:,}")
print(f"Year 4: {year4:,}")
print(f"Year 5: {year5:,}")
print("=" * 50)
````
