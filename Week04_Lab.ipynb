4. Create a function (or set of functions ) taking one parameter for a password.

The function (s) will print True if

  - the password contains at least one upper case letter,   
  - the password contains at least one lower case letter,
  - the password contains at least one digit and one special character like for example ‘@’.
- Extra (check if the password contains space. If so, print a suitable error message)

````python
'''
Threshold 4: Password Validator
Name: Emaan Fatima
MISIS: M01041173
'''

# Decorative Header
print('=' * 50)
print('Password Validator'.center(50))
print('=' * 50)

# Import module for symbols
import string

# Function to validate password
def password_validator(password):
    check_upper = check_lower = check_digit = check_special = False
    symbols = string.punctuation

    # Validate each character in the password
    for item in password:
        if item.isupper():
            check_upper = True
        elif item.islower():
            check_lower = True
        elif item.isdigit():
            check_digit = True
        elif item in symbols:
            check_special = True
        elif item.isspace():
            print("Space is not allowed")
            return False

    # All criteria met → valid
    if check_upper and check_lower and check_digit and check_special:
        return True
    else:
        return False

# Ask user for a password
user_password = input("Enter your password: ")

# Validate the password
if password_validator(user_password):
    print("Password is valid ✅")
else:
    print("Password is invalid ❌")
print('=' * 50)
````

9. (Optional)
Basic Regex for Pattern Matching
Import the re module and use it to search for an email address in a string.
Use regex to find all the words that start with a capital letter in a sentence.

````python
'''
Threshold 9: Regex Pattern Matching
Name: Emaan Fatima
MISIS: M01041173
'''

# Import Regex Module
import re

# Sample text
sample_text = "Hi Alice, my email is emaanedu.hw@gmail.com and ef434@gmail.edu. Do share this with Brian. Thanks! "

# Extract email addresses
email_matcher = r'[\w.%+-]+@[\w.-]+\.(?:com|org|net|edu|io)'
email_list = re.findall(email_matcher, sample_text)
print('=' * 60)
print(" - Emails found:\n", email_list)

# Extract capitalized words
capital_word_pattern = r'\b[A-Z]\w*\b'
capitalized_words = re.findall(capital_word_pattern, sample_text)
print(" - Capitalized words:\n", capitalized_words)
print('=' * 60)

````

12. Password Validation with Multiple Criteria: Ask the user for a password and validate whether it:
- Has more than 8 characters.
- Contains at least one uppercase letter.
- Contains at least one lowercase letter.
- Contains at least one number.
- Contains at least one special character (@, #, $, etc.).
- Extra: If the password contains a space, print an error message

````python
'''
Threshold 12: Password Validator
Name: Emaan Fatimaa
MISIS: M01041173
'''

# Decorative Header
print('=' * 50)
print('Password Validator'.center(50))
print('=' * 50)

# Import module for symbols
import string

# Function to validate password
def pass_validator(password):
    # Initialize all checks
    check_upper = check_lower = check_digit = check_special = False
    symbols = string.punctuation

    # Check length
    if len(password) <= 8:
        print("❌ Password length > 8.")
        return False

    # Check each character
    for elements in password:
        if elements.isupper():
            check_upper = True
        elif elements.islower():
            check_lower = True
        elif elements.isdigit():
            check_digit = True
        elif elements in symbols:
            check_special = True
        elif elements.isspace():           
            print("❌ No spaces allowed.")
            return False

    # Validate all criteria
    if check_upper and check_lower and check_digit and check_special:
        return True
    else:
        print("❌ Incomplete password requirements:\n"
              "- Uppercase letter\n"
              "- Lowercase letter\n"
              "- Digit\n"
              "- Special character (@, #, $, ...)")
        return False

# Ask user for a password
user_password = input("Enter your password: ")

# Validate the password
if pass_validator(user_password):
    print("Password is valid ✅")
print('=' * 50)
````

16. (Binary to hex) Write a function that parses a binary number and converts itinto a hex number. The function header is: def binaryToHex(binaryValue):
Write a test program that prompts the user to enter a binary number and displays the corresponding hexadecimal value.



````python
'''
Excellent 2: Binary to Hex Converter
Name: Emaan Fatima
MISIS: M01041173
'''

# Decorative Header
print('=' * 50)
print('Binary-Hex Converter'.center(50))
print('=' * 50)

# Function to convert binary to hexadecimal
def binaryToHex(binaryValue):
    # Convert binary to decimal using int()
    decimal_value = int(binaryValue, 2)
    # Compute hexadecimal
    hex_value = hex(decimal_value)[2:].upper()  # remove '0x' via slicing
    return hex_value

# Run program with error handling
try:
    binary_input = input("Enter a binary number: ")
    hex_value = binaryToHex(binary_input)
    print(f" - Binary {binary_input}: {hex_value} Hexadecimal.")
except Exception as e:
    print(f"Error: {e}")

print('=' * 50)
````
