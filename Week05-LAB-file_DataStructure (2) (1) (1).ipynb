### Activity 3: Temperature Logger

Write a program that:
1. Asks the user for temperatures for 5 days
2. Writes each temperature to "temps.txt" with the day number
   Format: "Day 1: 22°C"
3. Reads the file and calculates the average temperature

<hr>

`````python
'''
Threshold 3: Temperature Logger
Name: Emaan Fatima
MISIS: M01041173
'''

# Initilize lists
entries = []  
temp_list = []  

# Collect 5 days of temperatures
for day in range(1, 6):
    day_temp = float(input(f'Enter Day {day} Temperature: ')) 
    entries.append(f"Day {day}: {day_temp}°C\n")  # store line for file
    temp_list.append(day_temp)  # store numeric value

# Save 5 days of temperatures
with open("temps.txt", "w") as file:
    file.writelines(entries) 

# Read temperatures from file
with open("temps.txt", "r") as file:
    file_data = file.readlines()

# Calculate average temperature
avg_temp = sum(temp_list) / len(temp_list)

# Decorative Header
print("\n" + "=" * 30)
print("Temperature Logger".center(30))
print("=" * 30)

# Display temperatures
for record in file_data:
    print(record.strip())

# Display average temperature
print("=" * 30)
print(f"Average Temperature: {avg_temp:.2f}°C".center(30))
print("=" * 30)
`````

You are responsible for managing a study group for a course. Each student can be part of multiple study groups, and you need to track which groups a student is enrolled in. Students can join or leave study groups. What data structure would you choose to manage this efficiently, and why?

**Example data:**

```python
{"John": ["Math", "Physics"], "Emily": ["Physics", "Biology"], "Alex": ["Math"]}

Steps:

1. Choose a data structure that allows quick access to the groups each student belongs to.
2. Write a program that lets students join and leave study groups.
3. Implement a function to display all groups a student is part of.
<hr>

`````python
'''
Typical 15: Managing Study Groups
Name: Emaan Fatima
MISIS: M01041173
'''

'''
Answer: I am using a dictionary of lists in order to track students and their 
groups because it allows us to quickly access each student's groups 
and we can easily add/remove groups.
'''

# Study Group data
study_groups = {
    "John": ["Math", "Physics"],
    "Emily": ["Physics", "Biology"],
    "Alex": ["Math"]
}

# Function 1: Add student to a group
def add_to_group(student, group):
    study_groups[student].append(group)

# Function 1: Remove student from a group
def exit_group(student, group):
    study_groups[student].remove(group)

# Function 3: Show student's groups
def view_groups(student):
    for group in study_groups[student]:
        print(" -", group)

# Menu Loop
while True:
    print("\n--- Study Group Manager ---")
    print("📚 All study groups: 🧮 Math, ⚛️ Physics, 🧬 Biology")
    print("1. Join a group")
    print("2. Leave a group")
    print("3. Display student's groups")
    print("4. Exit")

    user_choice = input("Enter your choice (1-4): ")

    if user_choice == "1":
        student = input("Enter student name: ")
        group = input("Enter group name to join: ")
        add_to_group(student, group)
    elif user_choice == "2":
        student = input("Enter student name: ")
        group = input("Enter group name to leave: ")
        exit_group(student, group)
    elif user_choice == "3":
        student = input("Enter student name: ")
        view_groups(student)
    elif user_choice == "4":
        print("Exiting program. Bye!")
        break
    else:
        print("Error: Please try again.")
`````

### Activity 16: Managing and Shuffling Playlists

You are developing a music player that lets users create a playlist and shuffle it. Users should be able to add songs, remove songs, and shuffle the playlist randomly. What data structure would you use to represent the playlist, and why?

**Example data:**

```python
["Song A", "Song B", "Song C", "Song D"]
````

Steps:

1. Choose a data structure that allows easy addition, removal, and shuffling of songs.
2. Write a program that simulates playlist operations (add, remove, shuffle).
3. Let users shuffle the playlist and display the new order.
<hr>

`````python
'''
Typical 16: Managing and Shuffling Playlists
Name: Emaan Fatima
MISIS: M01041173
'''

'''
Answer: I'm using a list for the playlist because it allows us to
easily add, remove, and shuffle songs.
'''

# Import Random Module
import random

# Initialize playlist
my_playlist = ["Love Story", "Shake It Off", "Blank Space", "You Belong With Me", "Save Your Tears"]

# Add a song
def add_song(song):
    my_playlist.append(song)

# Remove a song
def remove_song(song):
    if song in my_playlist:
        my_playlist.remove(song)
    else:
        print('Song not Found!')

# Shuffle the playlist
def shuffle_songs():
    random.shuffle(my_playlist)

# Display playlist
def display_songs():
    print("\nPlaylist:")
    for song in my_playlist:
        print(f' - {song}')

# Menu Loop
while True:
    print('=' * 35)
    print("\tPlaylist Menu")
    print('=' * 35)
    print("1. Display playlist")
    print("2. Add song")
    print("3. Remove song")
    print("4. Shuffle playlist")
    print("5. Exit")
    print('=' * 35)
    
    user_choice = input("Enter your choice (1-5): ")
    
    if user_choice == "1":
        display_songs()
    elif user_choice == "2":
        new_song = input("Enter song name: ")
        add_song(new_song)
        print(f'"{new_song}" added.\n')
    elif user_choice == "3":
        erase_song = input("Enter song name: ")
        remove_song(erase_song)
        print(f'"{erase_song}" removed.\n')
    elif user_choice == "4":
        shuffle_songs()
        print("Playlist shuffled.\n")
    elif user_choice == "5":
        print("Exiting program.")
        break
    else:
        print("Invalid choice. Try again.\n")
`````

`````python
'''
Excellent 17: Counting Words in Les_Brown.txt
Name: Emaan Fatima
MISIS: M01041173
'''

# Read the file
with open("Les_Brown.txt", "r") as file:
    file_text = file.read()

# Split into words
word_list = file_text.split()

# Count occurrences via dictionary
word_count = {} 
for word in word_list:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1

# Save counts to file
with open("word_count.txt", "w") as new_file:
    for word, occurrences in word_count.items():
        new_file.write(f"{word}: {occurrences}\n")

print("Saving word counts to 'word_count.txt'...")
`````

### Activity 19 (Set): Splitting Clothing Stock by Size

Create a Python program that reads the `clothing_stock.txt` file and splits the clothing items into separate sets based on their size (e.g., all XS items in one set, all M items in another set). Print each set.

**Steps:**
1. Read the file and separate items by their size.
2. Store each size in a different set.
3. Print each set of items.
<hr>

`````python
'''
Excellent 19: Splitting Clothing Stock by Size
Name: Emaan Fatima
MISIS: M01041173
'''

# Read clothing items 
stock_set = set()
with open("clothing_stock.txt", "r") as file:
    for line in file:
        stock_set.add(line.strip())  # strip newline/extra spaces

# Store items by size
sizes = {"XXL": set(),
         "XL": set(),
         "L": set(), 
         "M": set(),
         "XS": set(),
         "S": set()}

# Split items by size
for item in stock_set: 
    for size in sizes:
        if item.endswith(size):
            sizes[size].add(item)
            break

# Decorative Header
print('=' * 110)
print('Stock by Size'.center(100))
print('=' * 110)

# Display items 
for size, items in sizes.items():
    print(f"\n - {size} Stock ({len(items)}):")
    print(", ".join(sorted(items)))
print('=' * 110)
`````

### Activity 21: Personal Finance App

You are creating a personal finance app that allows users to categorize their expenses and track spending by category (e.g., food, entertainment, travel). Users can add expenses, remove expenses, and see a summary of their spending. What data structure would you choose to store and organize the expenses, and why?

**Example data:**

```python
{"Food": [25, 40, 15], "Entertainment": [50, 20], "Travel": [300, 150]}
```
Steps:

1. Choose a data structure that lets users easily add and track expenses by category.
2. Write a program that allows users to input expenses, remove them, and view their total spending by category.
3. Implement a function to display a summary of all spending.
<hr>

`````python
'''
Excellent 21: Personal Finance App
Name: Emaan Fatima
MISIS: M01041173
'''

'''
I am using dictionaries because they allow easy categorization and quick access 
to each category's expenses.
'''

# Expenses dictionary 
expense_data = {
    "Food": [25, 40, 15],
    "Entertainment": [50, 20],
    "Travel": [300, 150],
    "Education": [200, 150] 
}

# Function 1: Add expense
def add_expense(expense_type, cost):
    if expense_type in expense_data:
        expense_data[expense_type].append(cost)
    else:
        expense_data[expense_type] = [cost]
    print(f"{expense_type}: ${cost} added.")

# Function 2: Remove expense
def remove_expense(expense_type, cost):
    if expense_type in expense_data:
        if cost in expense_data[expense_type]:
            expense_data[expense_type].remove(cost)
            print(f"{expense_type}: ${cost} removed.")
    
# Function 3: Display summary
def display_summary():
    print("\nExpense Summary:")
    for expense_type, costs in expense_data.items():
        total = sum(costs)
        print(f"{expense_type} | {costs} | Total: ${total}")

# Menu Loop
while True:
    print("\n--- 🏦 Personal Finance Menu ---")
    print("1️⃣ Add expense")
    print("2️⃣ Remove expense")
    print("3️⃣ View summary")
    print("4️⃣ Exit")
    
    user_input = input("Enter your choice (1-4): ")

    if user_input == "1":
        expense_type = input("Enter expense category: ")
        cost = float(input("Enter expense amount: "))
        add_expense(expense_type, cost)
    elif user_input == "2":
        expense_type = input("Enter expense category: ")
        cost = float(input("Enter expense amount: "))
        remove_expense(expense_type, cost)
    elif user_input == "3":
        display_summary()
    elif user_input == "4":
        print("Program End. Bye!")
        break
    else:
        print("Error: Invalid input!")

`````
