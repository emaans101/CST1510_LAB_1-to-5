1. **Password Strength Checker** 

**Scenario:**
You are a junior security analyst at a company. Employees often choose weak passwords, so you need to create a simple program to check password strength. Since you only know the basics of programming, your program will use simple rules.

**Requirements:**

1. Ask the user to enter a password.

2. If the password has fewer than 8 characters, print "Weak".

3. If the password has 8 or more characters, print "Moderate".

4. If the password has 12 or more characters, print "Strong"

````python
# Threshold 1: Password Strength Checker

# Define Function to check strength
def pass_strength_checker():

    # Ask user for name & password
    user_name = input("👤 Enter your name: ")
    user_password = input("🔑 Enter your password: ")

    # Measure password length
    password_length = len(user_password)

    # Displaying a welcome message
    print('-' * 50)
    print(f"\t\t🔐 Welcome, {user_name}!")
    print('-' * 50)

    # Lambda to check password strength
    password_strength = lambda length: "⚠️ Weak Password" if length < 8 else "🛡️ Strong Password" if length >= 12 else "✅ Moderate Password"

    print(f" - {password_strength(password_length)}")
    print('-' * 50)

# Calling Function
pass_strength_checker()
````

2. **Multiplication quiz**

Suppose you want to develop a program to help a first-grader practice multiplication. The program randomly generates two single-digit integers, number1 and number2, and displays to the student a question such as; What is 1 * 7. After the student types the answer, the program displays a message to indicate whether it is true or false.

To generate a random number, you can use the randint(a, b) function in the random module. This function returns a random integer i between a and b, inclusively. To obtain a random integer between 0 and 9, use randint(0, 9).

The program may be set up to work as follows:
Step 1: Randomly generate two single-digit integers for number1 (e.g., 4) and number 2 (e.g., 5)
Step 2: Prompt the student to answer, "What is 4 * 5?" (user input)
Step 3: Check whether the student’s answer is correct. 

````python
# Threshold 2: Multiplication Quiz

# Function for multiplication quiz
def multiplication_func():
    
    # Importing Random Module
    import random

    # Ask for user's name
    user_name = input('👤 Enter your name: ')

    # Generate two single-digit random numbers 
    number1 = random.randint(0, 9)
    number2 = random.randint(0, 9)
    correct_answer = (lambda a, b: a * b)(number1, number2)

    # Display welcome message
    print('-' * 50)
    print(f'\t\t 👋 Welcome, {user_name}!')
    print('-' * 50)

    # Ask multiplication question
    user_answer = int(input(f'🧮 What is {number1} * {number2}? '))

    # Check the answer
    if user_answer == correct_answer:
       print('\t✅ Correct Answer 🎉')
    else:
       print(f'\t❌ Incorrect Answer. The correct answer was {correct_answer}.')
    print('-' * 50)
    
# Calling Function
multiplication_func()
````

4. **Body mass index (BMI)** is a measure of health and weight.

It can be calculated by taking your weight in kilograms and dividing it by the square of your height in meters. The interpretation of BMI for people 20 years or older is as follows:


BMI	Interpretation
BMI < 18.5	Underweight
18.5 ≤ BMI ≤ 25.0	Normal
25.0 ≤ BMI ≤ 30.0	Overweight
30.0 ≤ BMI 	Obese
 
Write a program that prompts the user to enter weight in pounds and height in inches and display the BMI. Note that one pound is 0.45359237 kilograms, and one inch is 0.0254 meters


````python
# Threshold 4: Body Mass Index (BMI) Program

# Constants for conversion
POUND_TO_KG = 0.45359237
INCH_TO_METER = 0.0254

def bmi_calculator():
    # Ask user for name, weight, and height
    user_name = input("👤 Enter your name: ")
    weight_pounds = float(input(" - Enter your weight in pounds: "))
    height_inches = float(input(" - Enter your height in inches: "))

    # Convert units
    weight_kg = weight_pounds * POUND_TO_KG
    height_meter = height_inches * INCH_TO_METER

    # Calculate BMI
    bmi = weight_kg / (height_meter ** 2)

    # Welcome message
    print('-' * 50)
    print(f"\t\t👋 Welcome, {user_name}!")
    print('-' * 50)

    # BMI category
    bmi_category = lambda bmi: ("Underweight" if bmi < 18.5 else "Normal" if bmi <= 25.0 else "Overweight" if bmi <= 30.0 else "Obese")

    # Display BMI and category
    print(f" - BMI is: {bmi:.2f}")
    print(f" - BMI interpretation: {bmi_category(bmi)}")
    print('-' * 50)

# Call the function
bmi_calculator()
````

5. **Student Marks Analysis**
You are working as a data analyst in a school. You have the exam marks of students stored in a list. Your task is to analyze the data.

**Requirements**:

- Store marks of students in a list.

- Use a loop to calculate the average mark.

- Find the highest and lowest mark using list functions.

- Print the number of students who passed (marks ≥ 50) and failed.

````python
#Threshold 5: Student Marks Analysis

# Student marks
student_marks = [45, 62, 78, 50, 33, 89, 97, 58, 41, 76, 99]

# Initialize variables
total_marks = 0
pass_count = 0
fail_count = 0

# Sum marks and count pass & fail
for mark in student_marks:
    total_marks += mark         
    if mark >= 50:
        pass_count += 1        
    else:
        fail_count += 1        

# Calculating average mark
average_mark = total_marks / len(student_marks)

# Find highest and lowest marks
highest_mark = max(student_marks)
lowest_mark = min(student_marks)

# Print results
print('-' * 50)
print(f'\t🖊️ Student Marks Analysis!')
print('-' * 50)
print(f" - Average mark: {average_mark:.2f}")
print(f" - Highest mark: {highest_mark}")
print(f" - Lowest mark: {lowest_mark}")
print(f" • Number of students who passed: 🎉 {pass_count}")
print(f" • Number of students who failed: {fail_count}")
print('-' * 50)
````

As result, some people believe that it is better to count each of the first two human years as 10.5 dog years, and then count each additional human year as 4 dog years. 

Write a program that implements the conversion from human years to dog years described in the previous paragraph. 16. Write a **guessing game** in Python where the user has to guess a secret number. After every guess the program tells the user whether their number was too large or too small. At the end the number of tries needed should be printed.


````python
#Typical 5: Guessing Game

#Import Random Module
import random

# Number between 1 - 50
secret_number = random.randint(1, 50)

# Welcome message
print('-' * 50)
print("\t🎯 Guessing Game!")
print('-' * 50)
print("Guess a number between 1 and 50 ")

# Initializing variables
tries = 0 
user_guess = ''

# Looping until correct answer
while user_guess != secret_number:
    user_guess = int(input(" - Enter your guess: "))
    tries += 1  # count user tries

    # Check if guess is too high or too low
    if user_guess < secret_number:
        print("📉 Too small!")
    elif user_guess > secret_number:
        print("📈 Too large!")

# Correct Guess
print(f"🎉 Congratulations! You guessed it in {tries} tries.")
print('-' * 50)
````

17. **Collecting Positive Numbers**
You are writing a small program to help a teacher keep track of positive numbers entered by students. The teacher also wants to know if any even numbers were given.

**Requirements:**

1. Keep asking the user to enter a positive number, or type `"done"` to stop.
2. Store all entered numbers in a **list**.
3. After the input ends:

   * If numbers were entered:

     * Print all the numbers entered.
     * Check the list and print the **first even number**. If no even number exists, print `"No even numbers found"`.
   * If no numbers were entered, print `"No numbers were entered"`.


> ### Note 
>
>* **`lower()`**: Converts any text into lowercase.
>  Example: `"DONE".lower()` becomes `"done"`.
>  This makes it easier to compare user input without worrying about uppercase/lowercase.
>
>* **`isdigit()`**: Checks if a string only contains digits (0–9).
>  Example: `"123".isdigit()` returns `True`, but `"12a".isdigit()` returns `False`.
>  This helps us make sure the user typed a valid number before converting it into an integer.
> 


---

````python
# Typical 6: Collecting Positive Numbers

# Decorative output
print('-' * 50)
print("\t\t 🔢 Positive Numbers!")
print('-' * 50)

# Declare empty list
numbers = []

# Collect input until - 'done'
while True:
    user_input = input(" - Enter a positive number (enter 'done' to quit): ").lower()
    if user_input == 'done':
        break
    if user_input.isdigit() and int(user_input) > 0:
        numbers.append(int(user_input))
    else:
        print("🔴 Please enter a positive number.")

# Display numbers entered
if numbers:
    print(" - Numbers entered:", numbers)

    # Find first even number using filter & lambda
    even_number = None
    for num in filter(lambda x: x % 2 == 0, numbers):
        even_number = num
        break  # exit after finding first even number

    if even_number is not None:
        print('-' * 50)
        print(f"\t ➡️ First even number: {even_number}")
        print('-' * 50)
    else:
        print("No even numbers found")
else:
    print("No numbers were entered")
````

18. Suppose you want to develop a **program to play a lottery.**
    
The program randomly generates a two-digit number, prompts the user to enter a <u>two-digit number</u>, and determines whether the user wins according to the following rules:


    1. If the user’s input matches the lottery in the exact order, the award is £10,000.
    2. If all the digits in the user’s input match all the digits in the lottery number, the award is £3,000.
    3. If one digit in the user’s input matches a digit in the lottery number, the award is £1,000.



````python
# Excellent 1: Lottery Program 

# Import Random Module
import random

# Define Lottery Function
def play_lottery():
    # Generate random number
    random_num = str(random.randint(10, 99))

    print('=' * 45)
    print('\t\t 💰 Lottery Game')
    print('=' * 45)

    # Get user guess
    user_guess = input("Enter a two-digit number: ")

    print(f"🎲 Lottery number: {random_num}")

   # Determine Lottery based on guess
    if user_guess == random_num:
        print("🎉 You win £10,000!")
    elif sorted(user_guess) == sorted(random_num):
        print("🎉 You win £3,000!")
    elif user_guess[0] in random_num or user_guess[1] in random_num:
        print("🎉 You win £1,000!")
    else:
        print("No match. Better luck next time!")
    print('-' * 50)

# Call the Function to run the game
play_lottery()
````


21. **Password cracker** - write a program to print the accuracy of password guess.

For example: The actual password is `MDX123` while a user tried `MDX321` or `mdx123`, the program should tell that how accurately the password was guessed, in this case it should be `50%` accuracy. if a user enter `MDX124`, the output should be `83%` and in case of `123MDX` the output should be `0%`.

Points to keep in mind while calculating the accuracy a) case of the letters entered b) order of the letters

If a password is accurately guessed, the output should state `Password cracked!`

````python
# Excellent 4: Password Guessing Game

# Actual Password
correct_password = "MDX123"

# Decorative Print
print("=" * 50)
print("\t\tPASSWORD GUESSING GAME")
print("=" * 50)

# Inililizing tries
tries = 0

# Game loop
while True:
    user_guess = input("Enter your guess (enter 'exit' to quit): ")
    
    if user_guess.lower() == "exit":
        print("\nProgram Ended! Goodbye!")
        print("=" * 50)
        break

    tries += 1
    matches = 0

    # Password length
    password_length = len(correct_password)

    # Comparing password
    for i in range(password_length):
      if i < len(user_guess) and user_guess[i] == correct_password[i]:
        matches += 1


    # Calculating accuracy 
    password_accuracy = round((matches / password_length) * 100)

    print("-" * 50)  # Decorative printing

     # Checking if entire password is guessed
    if password_accuracy == 100:
        print(f"🎯 Well done! Password Guessed! ({tries} tries)")
        print("=" * 50)
        break
    else:
        # Showing accuracy 
        print(f"Nice try! Match score: {password_accuracy}%")
        print("Keep trying or type 'exit' to exit.\n")
````

````python
# Excellent 4: Password Guessing Game

# Actual Password
correct_password = "MDX123"

# Decorative Print
print("=" * 50)
print("\t\tPASSWORD GUESSING GAME")
print("=" * 50)

# Inililizing tries
tries = 0

# Game loop
while True:
    user_guess = input("Enter your guess (enter 'exit' to quit): ")
    
    if user_guess.lower() == "exit":
        print("\nProgram Ended! Goodbye!")
        print("=" * 50)
        break

    tries += 1
    matches = 0

    # Password length
    password_length = len(correct_password)

    # Comparing password
    for i in range(password_length):
      if i < len(user_guess) and user_guess[i] == correct_password[i]:
        matches += 1


    # Calculating accuracy 
    password_accuracy = round((matches / password_length) * 100)

    print("-" * 50)  # Decorative printing

     # Checking if entire password is guessed
    if password_accuracy == 100:
        print(f"🎯 Well done! Password Guessed! ({tries} tries)")
        print("=" * 50)
        break
    else:
        # Showing accuracy 
        print(f"Nice try! Match score: {password_accuracy}%")
        print("Keep trying or type 'exit' to exit.\n")
````
